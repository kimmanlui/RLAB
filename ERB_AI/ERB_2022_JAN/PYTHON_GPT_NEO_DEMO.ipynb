### 2.3 Build Your Own Function
Write up a function to include those lines.

```python
def gen_text(text):
  input_ids = tokenizer(text, return_tensors="pt").input_ids
  gen_tokens = model.generate(input_ids, do_sample=True, temperature=0.9, max_length=100,)
  gen_text = tokenizer.batch_decode(gen_tokens)[0]
  return(gen_text)
```

A sample text "**Welcome to the first day of class. This is the basic artificial intelligence course.**" is inputted. The AI will generate what might should come next. Note that "\n\n" means line break.

### 3.1 Sentence Completion

#### 3.1.1 Basic

```python
sentence            = "artificial intelligence is "
sentence_max_length = 15
do_sample           = True
# --------------- INPUT ---------------#
from happytransformer import HappyGeneration, GENSettings
happy_gen = HappyGeneration()  # the default model is GPT2
args      = GENSettings(max_length=sentence_max_length, do_sample=do_sample)
result    = happy_gen.generate_text(sentence, args=args) 
# --------------- OUTPUT --------------#
print("======")   
print(result.text)  #  a new field of research that has been gaining momentum in recent years.
```

#### 3.1.2 Put things together

```python
def gen_sentence_default(sentence,sentence_max_length, do_sample):
  from happytransformer import HappyGeneration, GENSettings
  happy_gen = HappyGeneration()  # the default model is GPT2
  args      = GENSettings(max_length=sentence_max_length, do_sample=do_sample)
  result    = happy_gen.generate_text(sentence, args=args) 
  return(result.text)
```

```python
gen_sentence_default("Enclosed is my resume." , 20, False )
```

#### 3.1.3 Change Pretrained Model

We can change the pre-trained model.

```python
def gen_sentence_neo(sentence,sentence_max_length, do_sample):
  from happytransformer import HappyGeneration, GENSettings
  happy_gen = HappyGeneration(model_type="GPT-NEO", model_name="EleutherAI/gpt-neo-125M")  # default uses gpt2
  args      = GENSettings(max_length=sentence_max_length, do_sample=do_sample)
  result    = happy_gen.generate_text(sentence, args=args) 
  return(result.text)
```

```python
gen_sentence_neo("Enclosed is my resume." , 20, False )
```

### 3.2 Sentiment Analysis*

```python
sentence="Well done."
# --------------- INPUT ---------------#
from happytransformer import HappyTextClassification
happy_tc = HappyTextClassification(model_type="DISTILBERT",  model_name="distilbert-base-uncased-finetuned-sst-2-english")
result = happy_tc.classify_text(sentence)
print(type(result))  # <class 'happytransformer.happy_text_classification.TextClassificationResult'>
print(result)        # TextClassificationResult(label='POSITIVE', score=0.9998761415481567)
# --------------- OUTPUT --------------#
print("======")
print("score:", str(result.score))
print("label:",result.label)  # LABEL_1
```

### 3.4 Question Answering

```python
sentence_list=["Today's date is January 10th, 2021", 
               "What is the date?"]
# --------------- INPUT ---------------#
from happytransformer import HappyQuestionAnswering
happy_qa = HappyQuestionAnswering()
result   = happy_qa.answer_question(sentence_list[0], sentence_list[1])
# --------------- OUTPUT --------------#
print("======")  
print(type(result))  # <class 'list'>
print(result)  # [QuestionAnsweringResult(answer='January 10th, 2021', score=0.9711642265319824, start=16, end=34)]
print(type(result[0]))  # <class 'happytransformer.happy_question_answering.QuestionAnsweringResult'>
print(result[0])  # QuestionAnsweringResult(answer='January 10th, 2021', score=0.9711642265319824, start=16, end=34)
print(result[0].answer)  # January 10th, 2021
```

### 3.5 Word Prediction
I think therefore I _____

```python
sentence="I think therefore I [MASK]"
# --------------- INPUT ---------------#
from happytransformer import HappyWordPrediction
happy_wp = HappyWordPrediction()  # default uses distilbert-base-uncased
result = happy_wp.predict_mask(sentence)
# --------------- OUTPUT --------------#
print("======") 
print(type(result))  # <class 'list'>
print(result)  # [WordPredictionResult(token='am', score=0.10172799974679947)]
print(type(result[0]))  # <class 'happytransformer.happy_word_prediction.WordPredictionResult'>
print(result[0])  # [WordPredictionResult(token='am', score=0.10172799974679947)]
print(result[0].token)  # am
print(result[0].score)  # 0.10172799974679947
```

### 3.6 Tag (Token Classification )

```python
sentence="My name is Geoffrey and I live in Toronto"
# --------------- INPUT ---------------#
from happytransformer import HappyTokenClassification
happy_toc = HappyTokenClassification(model_type="BERT", model_name="dslim/bert-base-NER")
result = happy_toc.classify_token(sentence)
# --------------- OUTPUT --------------#
print("======") 
print(type(result))  # <class 'list'>
print("how many tags:" + str(len(result)))
print(result[0].word)  # Geoffrey
print(result[0].entity)  # B-PER
print(result[0].score)  # 0.9988969564437866
print(result[0].index)  # 4
print(result[0].start) # 11
print(result[0].end)  # 19
print(result[1].word)  # Toronto
print(result[1].entity)  # B-LOC
```

### 3.7 Sentence Preduction
We would like to know if "How old are you?" logically comes with "I am 21 years old." 

```python
sentence_list=["How old are you?", 
               "I am 21 years old."]
# --------------- INPUT ---------------#
from happytransformer import HappyNextSentence
happy_ns = HappyNextSentence()
result = happy_ns.predict_next_sentence(
    sentence_list[0],
    sentence_list[1]
)
# --------------- OUTPUT --------------#
print("======") 
print(type(result))  # <class 'float'>
print(result)  # 0.9999918937683105
```
